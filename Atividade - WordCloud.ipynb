```python
#importando as bibliotecas necessárias
import os
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#definindo o caminho para o diretório onde se encontra o arquivo a ser analizado
os.chdir("/home/joaopaulo/Documentos/documentos pos graduacao/Atividade - WordCloud/")
```

```python
#armazenando o texto do arquivo em uma variavel
text = open('artigo.txt', "r").read()
```

```python
#definindo uma lista de stopwords
stopWords = set(stopwords.words('portuguese'))
```

```python
%matplotlib inline
```

```python
#plotando a wordcloud
plt.figure(figsize=(20,10))
wordcloud = WordCloud(background_color= 'white' , mode = 'RGB', width = 2000, height = 1000, stopwords = stopWords).generate(text)
plt.title("Nuvem de Palavras")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
